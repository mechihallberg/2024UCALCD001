```python
import pandas as pd
import seaborn as sb
from sklearn.linear_model import LinearRegression
```

```python
datos = pd.read_csv('datos.csv')
```

```python
datos.info()
```

```python
sb.scatterplot(x="celsius", y="fahrenheit", data=datos)
```

```python
sb.scatterplot(x="celsius", y="fahrenheit", data=datos, hue="fahrenheit")
```

```python
sb.scatterplot(x="celsius", y="fahrenheit", data=datos, hue="fahrenheit", palette="coolwarm")
```

```python
características = datos["celsius"]
etiquetas = datos["fahrenheit"]
```

```python
type(características)
type(características.values)
```

```python
características.values
```

```python
características_procesada = características.values.reshape(-1, 1)
etiquetas_procesada = etiquetas.values.reshape(-1, 1)
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(características_procesada, etiquetas_procesada)
```

```python
grados_celsius = float(input("Ingrese un valor en grados Celsius: "))
```

```python
predicción = modelo.predict([[grados_celsius]])
```

```python
modelo.score(características_procesada, etiquetas_procesada)
```

```python
print(f"{grados_celsius} grados celcius son {predicción} grados fahrenheit")
```
